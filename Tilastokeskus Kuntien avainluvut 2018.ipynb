```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Tyyli vaikuttaa grafiikan ulkoasuun
# Käytettävissä olevat tyylit voit listata komennolla plt.style.available
plt.style.use('seaborn-whitegrid')
```

```python
# Korrelaatiokertoimet
df.corr()
```

```python
# Seaborn-kirjaston heatmap väritää korrelaatiokertoimia
sns.heatmap(df.corr(), annot = True)
```

```python
# Korrelaatioiden visuaalinen tarkastelu hajontakaavioina
# Pairplotin laskeminen näin monesta muuttujasta kestää hetken aikaa!
sns.pairplot(df, kind = 'reg')
```

```python
# Piirrä kolme kuvaajaa: asukaat-huoltosuhde, asukkaat-työllisyysaste ja huoltosuhde-työllisyysaste
# Yksittäisen muuttujaparin tarkastelu hajontakaaviona
# Mitä korkeampi huoltosuhde, sitä matalampi väkiluku
ax1 = sns.jointplot(data = df, x = 'Väkiluku', y = 'Taloudellinen huoltosuhde', kind = 'reg')

# Akselien skaalaus
ax1.ax_marg_x.set_xlim(50000, 700000)
ax1.ax_marg_y.set_ylim(50, 300)
```

```python
# Piirrä kuvaajaa: asukkaat-työllisyysaste
# Yksittäisen muuttujaparin tarkastelu hajontakaaviona
# Mitä korkeampi väkiluku, sitä korkeampi työllisyysaste
ax1 = sns.jointplot(data = df, x = 'Väkiluku', y = 'Työllisyysaste, %', kind = 'reg')

# Akselien skaalaus
ax1.ax_marg_x.set_xlim(50000, 700000)
ax1.ax_marg_y.set_ylim(60, 100)
```

```python
# Piirrä kolme kuvaajaa:huoltosuhde-työllisyysaste
# Yksittäisen muuttujaparin tarkastelu hajontakaaviona
# Mitä korkeampi huoltosuhde, sitä matalampi työllisyysaste
ax1 = sns.jointplot(data = df, x = 'Taloudellinen huoltosuhde', y = 'Työllisyysaste, %', kind = 'reg')

# Akselien skaalaus
ax1.ax_marg_x.set_xlim(50, 300)
ax1.ax_marg_y.set_ylim(60, 100)
```
